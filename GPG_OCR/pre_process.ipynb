```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pytesseract
```

```python
plt.rcParams['figure.figsize'] = (8,12)
img = plt.imread("Data/cont1.jpg")
plt.imshow(img)
plt.show()
```

```python
img_bw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_bw *=255
plt.imshow(img_bw, cmap='binary')
plt.show()
```

```python
ret ,img_thr = cv2.threshold(img_bw,60,255,cv2.THRESH_BINARY_INV)
plt.imshow(img_thr, cmap='gray')
plt.show()
```

```python
text = pytesseract.image_to_string(img_thr)
print(text)
```

```python
#blur
img_bl = cv2.blur(img_thr, (25,25))
plt.imshow(img_bl, cmap="gray")
plt.show()
```

```python
ret ,img_thr = cv2.threshold(img_bl,50,255,cv2.THRESH_BINARY_INV)
plt.imshow(img_thr, cmap='gray')
plt.show()
```

```python
from matplotlib import image


img2 = np.asarray(img_thr, dtype=np.uint8)
contours, hie = cv2.findContours(img2, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

img3 = cv2.cvtColor(img_thr,cv2.COLOR_GRAY2RGB)

# path = "/opt/homebrew/Cellar/tesseract/5.2.0/"
# pytesseract.tesseract_cmd = path

for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    if (w<10) or (h<10):
        continue
    cv2.rectangle(img3,(x,y), (x+w,y+h), (0,255,0),2)
    croped = img_bw[y:y+h, x:x+w]
    text = pytesseract.image_to_string(croped)
    print(text)
```
