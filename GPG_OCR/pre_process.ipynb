```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (8,12)
img = plt.imread("Data/test.png")
plt.imshow(img)
plt.show()
```

```python
img_bw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_bw *=255
plt.imshow(img_bw, cmap='binary')
plt.show()
```

```python
ret ,img_thr = cv2.threshold(img_bw,192,255,cv2.THRESH_BINARY_INV)
plt.imshow(img_thr, cmap='gray')
plt.show()
```

```python
#blur
img_bl = cv2.blur(img_thr, (40,40))
plt.imshow(img_bl, cmap="gray")
plt.show()
```

```python
img2 = np.asarray(img_bl, dtype=np.uint8)
contours, hie = cv2.findContours(img2, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

img2 = cv2.cvtColor(img2,cv2.COLOR_GRAY2RGB)

for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    if (w<10) or (h<10):
        continue
    cv2.rectangle(img,(x,y), (x+w,y+h), (0,255,0),2)
    # croped = img_thr[y:y+h, x:x+w]
```
